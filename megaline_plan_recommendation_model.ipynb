```python
import pandas as pd
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

### Abrir y leer los datos de el Dataset

```python
df = pd.read_csv('/datasets/users_behavior.csv')
df.info()
```

```python
df.head(10)
```

### Se hizo el segmento de los datos en el conjunto
Como la columna 'is_ultra' es : plan para el mes actual (Ultra - 1, Smart - 0), lo que vamos a hacer es un modelo de clasificaci√≥n. 
- por lo que se separ√≥ los datos (features) del target (Ultra - 1, Smart - 0)
- Adem√°s de eso escogimos el 25% de los datos para la validaci√≥n

```python
features = df.drop('is_ultra', axis=1)
target = df['is_ultra']

features_train, features_valid, target_train, target_valid = train_test_split(
    features, target, test_size=0.25, random_state=12345)
```

### Se investiga la calidad de diferentes modelos cambiando los hiperpar√°metros.

```python
best_score = 0
best_est = 0
for est in range(1, 51): # selecciona el rango del hiperpar√°metro
    model = RandomForestClassifier(random_state=54321, n_estimators=est) # configura el n√∫mero de √°rboles
    
    model.fit(features_train, target_train) # entrena el modelo en el conjunto de entrenamiento
    score = model.score(features_valid, target_valid) # calcula la puntuaci√≥n de accuracy en el conjunto de validaci√≥n
    if score > best_score:
        best_score = score# guarda la mejor puntuaci√≥n de accuracy en el conjunto de validaci√≥n
        best_est = est# guarda el n√∫mero de estimadores que corresponden a la mejor puntuaci√≥n de exactitud

print("La exactitud del mejor modelo en el conjunto de validaci√≥n (n_estimators = {}): {}".format(best_est, best_score))
```

#### Se encontr√≥ la mejor opcion en cuanto a los hiperparametros 
La exactitud del mejor modelo en el conjunto de validaci√≥n (n_estimators = 28): 0.7972636815920398
- Entonces se hizo el modelo final con los datos de los hiperparametros que encontramos

```python
model = RandomForestClassifier(random_state=54321, n_estimators=28) # configura el n√∫mero de √°rboles

model.fit(features_train, target_train)
```

#### Se hizo la predicci√≥n con los datos de validaci√≥n 
- Se imprimi√≥ el error 

```python
score = model.score(features_valid, target_valid)
print("La exactitud del mejor modelo en el conjunto de validaci√≥n",score)
```

###  Usa el dataset para comprobar la exactitud.

```python
predictions_final = model.predict(features)

def error_count(answers, predictions):
    count = 0
    for i in range(len(answers)):
        if answers[i] != predictions[i]:
            count += 1
    return count

def accuracy(answers, predictions):
    err = error_count(answers, predictions)
    exactitud = (len(answers)-err)/len(answers)
    return exactitud
    

print('Accuracy:', accuracy(target, predictions_final))
```
